```python
import pandas as pd
```

```python
train_df = pd.read_csv('../data/train.csv')
```

```python
train_df.head()
```

```python
train_df.drop(columns=['state','account_length','area_code'], inplace= True)
```

```python
train_df.head()
```

```python
train_df.isnull().sum()
```

```python
train_df.describe()
```

```python
train_df['international_plan'].replace({'no':0,'yes':1}, inplace=True)
train_df['voice_mail_plan'].replace({'no':0,'yes':1}, inplace=True)
train_df['churn'].replace({'no':0,'yes':1}, inplace=True)
```

```python
train_df.head()
```

```python
train_df['minutes'] = train_df['total_day_minutes'] + train_df['total_eve_minutes'] + train_df['total_night_minutes']
train_df['calls'] = train_df['total_day_calls'] + train_df['total_eve_calls'] + train_df['total_night_calls']
train_df['charge'] = train_df['total_day_charge'] + train_df['total_eve_charge'] + train_df['total_night_charge']
```

```python
train_df.head()
```

```python
train_df.drop(columns=['total_day_minutes','total_eve_minutes','total_night_minutes'],inplace=True)
train_df.drop(columns=['total_day_calls','total_eve_calls','total_night_calls'],inplace=True)
train_df.drop(columns=['total_day_charge','total_eve_charge','total_night_charge'],inplace=True)
```

```python
train_df.head()
```

```python
train_df.drop(columns=['total_intl_minutes','total_intl_calls','total_intl_charge'],inplace=True)
```

```python
train_df.head()
```

```python
y = train_df["churn"]
```

```python
X = train_df.drop("churn", axis = 1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Metric:  F1 score, Accuracy, Precision, Recall, ... 

from sklearn.metrics import f1_score, accuracy_score
```

```python
f1 = f1_score(y_test, y_pred)
acc = accuracy_score(y_test,y_pred)
```

```python
f1,acc
```

```python
# Accuracy =  Tong so lan du doan dung / tong so sample
```

```python
import mlflow
from urllib.parse import urlparse

mlflow.set_tracking_uri('http://localhost:1234')
mlflow.set_experiment('CustomerChurn01')

with mlflow.start_run():
    model = RandomForestClassifier()
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    f1 = f1_score(y_test, y_pred)
    acc = accuracy_score(y_test,y_pred)

    mlflow.log_metric("F1 Score",f1)
    mlflow.log_metric("Accuracy Score", acc)

    tracking_url_type_store = urlparse(mlflow.get_artifact_uri()).scheme

    if tracking_url_type_store != "file":
        mlflow.sklearn.log_model(model,"model",registered_model_name="CustomerChurnModel")
    else:
        mlflow.sklearn.load_model(model)
```
